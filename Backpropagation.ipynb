```python
import math

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

# Derivative of sigmoid
def sigmoid_derivative(y):
    return y * (1 - y)

# Inputs
x1 = 0.35
x2 = 0.9

# Target output
t = 0.5

# Learning rate
eta = 1

# Initial weights (from image)
w13 = 0.1
w14 = 0.4
w23 = 0.8
w24 = 0.6
w35 = 0.3
w45 = 0.9

# --------------------
# FORWARD PROPAGATION
# --------------------

# Hidden layer
net_h3 = x1 * w13 + x2 * w23
y3 = sigmoid(net_h3)

net_h4 = x1 * w14 + x2 * w24
y4 = sigmoid(net_h4)

# Output layer
net_o5 = y3 * w35 + y4 * w45
y5 = sigmoid(net_o5)

print("Forward Propagation Output:")
print("y3 =", y3)
print("y4 =", y4)
print("y5 =", y5)

# --------------------
# BACKPROPAGATION
# --------------------

# Output layer error
error = t - y5
delta_o5 = error * sigmoid_derivative(y5)

# Update weights (Hidden → Output)
w35_new = w35 + eta * delta_o5 * y3
w45_new = w45 + eta * delta_o5 * y4

# Hidden layer error
delta_h3 = sigmoid_derivative(y3) * delta_o5 * w35
delta_h4 = sigmoid_derivative(y4) * delta_o5 * w45

# Update weights (Input → Hidden)
w13_new = w13 + eta * delta_h3 * x1
w23_new = w23 + eta * delta_h3 * x2

w14_new = w14 + eta * delta_h4 * x1
w24_new = w24 + eta * delta_h4 * x2

print("\nUpdated Weights After Backpropagation:")
print("w13 =", w13_new)
print("w23 =", w23_new)
print("w14 =", w14_new)
print("w24 =", w24_new)
print("w35 =", w35_new)
print("w45 =", w45_new)
```
